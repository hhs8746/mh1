```python
#  
#     0.5 
p=0.5 
  
# 0 
start = 0
rand_walks = [start]

# 0 1  500 
t = np.random.random(500)
down = t < p
up = t >= p

#  
# z(t) = z(t-1) + a(t) 
for d, u in zip(down, up):
    rand_walks.append(rand_walks[-1] - d + u)

#  
plt.plot(rand_walks)
plt.show()
```
